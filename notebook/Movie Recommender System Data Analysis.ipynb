## Content Based Recomender System

```python
import pandas as pd
import numpy as np
```

```python
credit = pd.read_csv(r"..\archive\tmdb_5000_credits.csv")
credit
```

```python
movies = pd.read_csv(r"../archive/tmdb_5000_movies.csv")
movies
```

```python
credit.shape, movies.shape
```

```python
final = movies.merge(credit, on='title')
final.head()
```

```python
final.shape
```

```python
# Keeping important columns for recommendation
final = final[['movie_id','title','overview','genres','keywords','cast','crew']]
final.head(2)
```

```python
final.isnull().sum()
```

```python
final.dropna(inplace=True)
```

```python
final.isnull().sum()
```

```python
final.duplicated().sum()
```

```python
# Handeling The Genere Column
final.iloc[1]['genres']
```

```python
import ast # Converting The String To List

def convert_list(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
        
    return L
```

```python
final['genres'] = final['genres'].apply(convert_list)
```

```python
final.head(2)
```

```python
# Handeling The keywords Column
final.iloc[1]['keywords']
```

```python
final['keywords'] = final['keywords'].apply(convert_list)
```

```python
final.head(2)
```

```python
# Handeling The Cast Column
final.iloc[0]['cast']
```

```python
# Here I am Slecting Top 4 Cast
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 4:
            L.append(i['name'])
        counter += 1
    return L
```

```python
final['cast'] = final['cast'].apply(convert_cast)
```

```python
final.head(2)
```

```python
# Handeling The Crew Column
final.iloc[0]['crew']
```

```python
# Fetch Only Director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
final['crew'] = final['crew'].apply(fetch_director)
```

```python
final.head(2)
```

```python
# Handeling The Overview Column
final.iloc[0]['overview']
```

```python
final['overview'] = final['overview'].apply(lambda x:x.split())
final.sample(4)
```

```python
# Now Removing space like that 
'Anna Kendrick'
'AnnaKendrick'

def remove_space_between(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
final['cast'] = final['cast'].apply(remove_space_between)
final['crew'] = final['crew'].apply(remove_space_between)
final['genres'] = final['genres'].apply(remove_space_between)
final['keywords'] = final['keywords'].apply(remove_space_between)
```

```python
final.sample(4)
```

```python
# Concatenate
final['tags'] = final['overview'] + final['keywords'] + final['genres'] + final['cast'] + final['crew']
final.head()
```

```python
final.iloc[0]['tags']
```

```python
new_df = final[['movie_id', 'title', 'tags']]
new_df.head()
```

```python
# Connvert List To Str
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
# Connvert Lower Case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()

def stems(text):
    T = []
    for i in text.split():
        T.append(ps.stem(i))
        
    return " ".join(T)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector
```

```python
vector.shape, len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
index=new_df[new_df['title'] == 'The Lego Movie'].index[0]
index
```

```python
distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
distance
```

```python
for i in distance[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distance[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man 2')
```
